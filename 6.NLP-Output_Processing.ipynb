# Find Frequently Appeared Name

```python
def Find_Common(df,number):
    df_com = df.Organization.str.get_dummies(sep=' ').sum()
    df_com = pd.DataFrame(df_com.sort_values(ascending=False))
    df_com = df_com.reset_index()
    df_com = df_com.rename(columns={0:"Count","index":"Tag"})
    df_com = df_com.loc[df_com["Count"]>=number]
    return df_com
```

# Refine the identification of the tag

```python
def Refine_Overall(df):
    #United the tag
    special =['ARCO','ITRIA','PAYPAL','UBER','VENMO','ZELLE']
    
    automobile=[ 'ADVANCE AUTO','CAPITAL ONE AUTO','CIRCLE K','CHEVRON', 'CITGO',\
             'DAT SOLUTIONS','MURPHY EXPRESS', 'EXXONMOBIL', 'KROGER',\
             'MARATHON','MAVERIK','OREILLY AUTO', 'PILOT',\
             'RACETRAC', 'SHELL', 'SPEEDWAY', 'SUNOCO',\
             'TEXACO', 'THORNTONS', 'TRACTORS']
    
    insurance=['AAA','AETNA','AFLAC','ALLIANZ','AM INCOME','AMTRUST','ANTHEM BLUE','ALLSTATE','GEICO','GLOBE LIFE','HARTFORD','HISCOX','HUMANA',\
               'IPFS','LIBERTY MUTUAL','NATIONAL LIFE','NATIONAL GENERAL','NATIONWIDE','NORTHWESTERN','PEACOCK','PRIMERICA',\
               'STATE FARM','TRANSAMERICA','TRUEBILL','USAA','WESTGUARD' ]
    
    retail = ['BEST BUY','STRIPE','TARGET','WALMART']
    
    
    
    for i,j in df.iterrows():
        flag = False
        try:
            if re.search(r"PAY(.*?)X",j["Organization"]):
                df.at[i,["Organization"]] = "PAYCHEX"
            elif re.search(r"A(.*?)M(.*?)Z",j["Organization"]):
                df.at[i,["Organization"]] = "AMAZON"
            elif re.search(r"HOME(.*?)DEPOT",j["Organization"]):
                df.at[i,["Organization"]] = "HOME DEPOT"
            elif re.search(r"\bWM\b",j["Organization"]):
                df.at[i,["Organization"]] = "WALMART"
            else:        
                for k in special:
                    if re.search(r"\b"+k,j["Organization"]):
                        df.at[i,["Organization"]] = k
                        flag = True
                        break
                if flag == False:
                    for k in automobile:
                        if re.search(k,j["Organization"]):
                            df.at[i,["Organization"]] = k
                            flag = True
                            break
                if flag == False:
                    for k in insurance:
                        if re.search(k,j["Organization"]):
                            df.at[i,["Organization"]] = k
                            flag = True
                            break
                if flag == False:
                    for k in retail:
                        if re.search(k,j["Organization"]):
                            df.at[i,["Organization"]] = k
                            flag = True
                            break
        except:
            print(j["Description"])
            

    return df
```

# Groupby Organization name

```python
def Groupby_Org(df):
    df_gp = df.groupby(["Organization"]).agg({"Organization":["count"]})
    df_gp.columns = df_gp.columns.droplevel()
    df_gp= df_gp.reset_index()
    df_gp = df_gp.sort_values("count",ascending=False)
    return df_gp
```
