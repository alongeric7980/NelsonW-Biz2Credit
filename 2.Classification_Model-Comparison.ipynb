```python
#Function for defining the type of attributes from Group or Class
def Ct_Per(df):
    for i,j in df.iterrows():
        if j[0]==j[1]==j[2]:
            df.loc[i, "Type" ] = "Full"
        elif j[0]==j[1] or j[0]==j[2]:
            df.loc[i, "Type" ] = "Half"
        else:
            df.loc[i, "Type" ] = "Inconsistent"
    return df
```

```python
def Final_Table(df_table):
    for i,j in df_table.iterrows():
        df_table.loc[i,j[3]]= j[0]+j[1]+j[2]
    df_table = df_table.rename(columns={0.0 : "Total Case"})
    df_table = df_table.drop(columns={"T"})
    df_table = df_table.astype('int32')
    df_table["Full Percentage"]= (df_table["Full"]/df_table["Total Case"])*100
    df_table["Half Percentage"]= (df_table["Half"]/ df_table["Total Case"])*100
    df_table["Inconsistent Percentage"] = (df_table["Inconsistent"]/df_table["Total Case"])*100
    df_table = df_table.reindex(sorted(df_table.columns), axis=1)
    df_table.drop(df_table[df_table["Total Case"]<50].index.values, axis=0, inplace=True)
    df_table.sort_values(by = "Full Percentage", ascending=False,inplace=True)
    return df_table 
```

# Class

```python
df_class = pd.DataFrame()
df_class["classification orig"] = df["classification orig"]
df_class["classification v2"] = df["classfication v2"]
df_class["classification hybrid"] = df["classification hybrid"]
df_class["transaction_type"] = df["transaction_type"]
df_class["amount"] = df["amount"]
df_class["org_biz_cat_id"] = df["org_biz_cat_id"]
df_class["biz_cat_id_v2"] = df["biz_cat_id_v2"]
df_class["biz_Cat_id_hybrid"] = df["biz_Cat_id_hybrid"]
df_class["Type"]="Full"
df_class["Description"]=df["description"]
```

```python
#Transform all value that contains "Transfer" into the same value
df_class.loc[df_class['classification orig'].str.contains('Transfer'), 'classification orig'] = 'Transfers'
df_class.loc[df_class['classification v2'].str.contains('Transfer'), 'classification v2'] = 'Transfers'
df_class.loc[df_class['classification hybrid'].str.contains('Transfer'), 'classification hybrid'] = 'Transfers'
df_class
```

```python
%%time
class_type = Ct_Per(df_class)
class_type
```

### Class Orig

```python
#class_orig
class_orig_gp = class_type.groupby([class_type.columns[0],"Type"]).agg({"Type":["count"]})
class_orig_gp.columns = class_orig_gp.columns.droplevel()
class_orig_gp= class_orig_gp.reset_index()
#class_orig_gp
class_orig_table = class_orig_gp.pivot(index=[class_orig_gp.columns[0]], columns='Type', values='count')
class_orig_table = class_orig_table.fillna(0)
class_orig_table["T"]=0

class_orig_final = Final_Table(class_orig_table)
class_orig_final
```

```python
#class_v2
class_v2_gp = class_type.groupby([class_type.columns[1],"Type"]).agg({"Type":["count"]})
class_v2_gp.columns = class_v2_gp.columns.droplevel()
class_v2_gp= class_v2_gp.reset_index()
class_v2_gp

class_v2_table = class_v2_gp.pivot(index=[class_v2_gp.columns[0]], columns='Type', values='count')
class_v2_table = class_v2_table.fillna(0)
class_v2_table["T"]=0
class_v2_table

class_v2_final = Final_Table(class_v2_table)
class_v2_final
```

### Class Hybrid

```python
#class_hybrid
class_hybrid_gp = class_type.groupby([class_type.columns[2],"Type"]).agg({"Type":["count"]})
class_hybrid_gp.columns = class_hybrid_gp.columns.droplevel()
class_hybrid_gp= class_hybrid_gp.reset_index()
class_hybrid_gp

class_hybrid_table = class_hybrid_gp.pivot(index=[class_hybrid_gp.columns[0]], columns='Type', values='count')
class_hybrid_table = class_hybrid_table.fillna(0)
class_hybrid_table["T"]=0
class_hybrid_table

class_hybrid_final = Final_Table(class_hybrid_table)
class_hybrid_final
```
